### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
# Object Oriented approach; Explicit, gives me more control and flexibility as I directly create the figues and axes and can customize the plot. 
# Pyplot approach; Implicit, the figures and axes are created "behind the scenes" for me. Great for simple and quick plots. 
# The recommended approach is the Object Oriented one, especially for reusable and/ or complex plots.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# Figure: The Figure refers to the Whole figure for everything in a plot. Can contain multiple axes. 
# Axes: This is the area (can be plural) where data is plotted. Including x and y axis within the axes (or more axes for more dimensions), titels and labels.
# Axis: Refers to the singular/ individual x and y axis within the axes object. 
# Artist: Is everything thats visible in a plot; including lines, texts, etc. 
```

### When plotting in Matplotlib, what is the expected input data type?

```python
# The expected input data type in Matplotlib is np.array() or numpy.ma.masked.array() 
# In other cases we can convert the object to arrays using the numpy.asarray()
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 100)
y = x**2

# Object Oriented approach 
fig, ax, = plt.subplots()
ax.plot(x, y)
ax.set_title("Object Oriented y = x²")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

```python
# Pyplot approach
plt.plot(x, y)
plt.title("Pyplot y = x²")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))

#ax1 scatter plot
ax1.scatter(random_data_x, random_data_y)
ax1.plot(x, y)
ax1.set_title("Scatter Plot")
ax1.set_xlabel("x")
ax1.set_ylabel("y")

#ax2 bar plot
ax2.bar(names, values)
ax2.set_title("How Many Fruits")
ax2.set_xlabel("Fruit")
ax2.set_ylabel("Amount")
```
